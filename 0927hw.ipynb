```python
import pandas as pd

fn = 'extended_csvReport.csv'

df = pd.read_csv(fn)
print(df)
```

```python
print('各地區不同年代銷售總額：')
totalSales =  df.groupby(['Year','Location'])['Revenue'].sum().reset_index()
print(totalSales)
```

```python
print('各地區2015-2024年銷售總額:')
totalSales2015_2024 = totalSales.groupby(['Location'])['Revenue'].sum().reset_index()
print(totalSales2015_2024)
```

```python
print('銷售總額最高的區域:')
bestSales = totalSales2015_2024.sort_values(by='Revenue', ascending=False).head(1)
print(bestSales)
```

```python
print('產品在該地區銷售總數量:')
totalQuantity = df.groupby(['Location','Product'])['Quantity'].sum()
print(totalQuantity)
```

```python
print('最受歡迎的產品:')
popularProduct = totalQuantity.loc[totalQuantity.groupby('Location').idxmax()]
print(popularProduct)
```

```python
print('業務員年均銷售額')
average = df.groupby(['Name'])['Revenue'].mean().reset_index()
print(average)
```

```python
print('績效最佳的業務員:')
bestaverage = average.sort_values(by='Revenue', ascending=False).head(1)
print(bestaverage)
```

```python
print('業務員銷售的商品數量:')
sale = df.groupby(['Name','Product'])['Quantity'].sum().reset_index()
print(sale)
```

```python
print('業務員最擅長銷售的茶葉:')
bestSale = sale.loc[sale.groupby('Name')['Quantity'].idxmax()]
print(bestSale)
```

```python
print('各產品每年銷售額:')
productSales = df.groupby(['Year','Product'])['Revenue'].sum().reset_index()
print(productSales)
```

```python
print('各產品每年成長率:')
productSales['Revenue Growth'] = productSales.groupby('Product')['Revenue'].pct_change()
print(productSales)
```

```python
print('銷售額增長最快的產品')
fastGrowth = productSales.sort_values(by='Revenue Growth', ascending=False).head(1)
print(fastGrowth)
```

```python
print('銷售額下降的產品')
slowGrowth = productSales.sort_values(by='Revenue Growth', ascending=True).head(1)
print(slowGrowth)
```
