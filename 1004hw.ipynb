```python
import requests
from bs4 import BeautifulSoup
import csv

url = 'http://fchart.github.io'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'lxml')
```

```python
imgtags = soup.find_all('img')
immgtagsData = [{'index': i, 'src': img.get('src')} for i, img in enumerate(imgtags)]
with open('imgtags.csv', 'w', newline = '',encoding='utf-8') as csvfile:
    fieldnames = ['index', 'src']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for row in immgtagsData:
        writer.writerow(row)
print('圖片標籤解析完成，已另存成CSV檔')
```

```python
paragraph = soup.find_all('p')
editorParagraph = [p.get_text() for p in paragraph if '編輯器' in p.get_text()]
editorParagraph.sort(key=len, reverse=True)
editorData = [{'index': i, 'paragraph': para} for i, para in enumerate(editorParagraph)]
with open('editorParagraph.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['index', 'paragraph']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for data in editorData:
        writer.writerow(data)
print('編輯器解析完成，已另存成CSV檔')
```
