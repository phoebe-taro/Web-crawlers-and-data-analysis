```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime

url = "https://www.ettoday.net/"
response = requests.get(url)
response.encoding = 'utf-8'

soup = BeautifulSoup(response.text, "lxml")
news_list = soup.find("div", class_="part_list_8")

data = []
for item in news_list.find_all("div", class_="piece"):
  category = item.find("em").text.strip() if item.find("em") else ""
  title = item.find("a").text.strip() if item.find("a") else ""
  news_url = item.find("a")['href'] if item.find("a") else ""
  date = item.find("span", class_="date").text.strip() if item.find("span", class_="date") else ""
  data.append([category, title, news_url, date])

for news_item in data:
  print(f"新聞類別: {news_item[0]}")
  print(f"新聞標題: {news_item[1]}")
  print(f"URL網址: {news_item[2]}")
  print(f"時間: {news_item[3]}")
  print("-" * 50)

df = pd.DataFrame(data,columns=["新聞類別", "新聞標題", "URL網址", "時間"])
df['時間'] = df['時間'].apply(lambda x: datetime.strptime(x,"%m-%d %H:%M")).dt.strftime('%m-%d %H:%M')
df = df.sort_values(by=['新聞類別', '時間'],ascending=[True,False])

df.to_csv('即時新聞解析清單.csv', index=False)
print('即時新聞解析清單.csv建立完成')
```
