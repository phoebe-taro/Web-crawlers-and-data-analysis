```python
print(' 年份   差額  成交價格 是否小於')
import csv

fn = 'stock1101.csv'

with open(fn,'r',encoding = 'big5') as fn0bj:
    reader = csv.reader(fn0bj)
    next(reader)

    for row in reader:
        diff = float(row[4]) - float(row[6])
        price = int(row[2].replace(',','')) / int(row[1].replace(',',''))
        less = price < float(row[8])
        print(f'{row[0]:>3}年  {diff:>5.2f}   {price:4.2f}   {less:} ')
```
